## Estandarizar Unidades

### Estandarizar unidades a d√≠as
> CI_Age at Adverse & Event CI_Age Unit

```python
data['age_in_days'] = np.where(data['CI_Age Unit'] == 'Year(s)', data['CI_Age at Adverse Event']*365,
                    np.where(data['CI_Age Unit'] == 'Month(s)', data['CI_Age at Adverse Event']*30.44,
                    np.where(data['CI_Age Unit'] == 'Week(s)', data['CI_Age at Adverse Event']*7,
                    np.where(data['CI_Age Unit'] == 'Day(s)', data['CI_Age at Adverse Event'],
                    np.where(data['CI_Age Unit'] == 'Decade(s)', data['CI_Age at Adverse Event']*3650,
                    np.where(data['CI_Age Unit'] == 'Not Available', np.nan,
                    data['CI_Age at Adverse Event']))))))
```
